## In-degree and Out-degree distributions

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df1 = pd.read_csv('in-degree_dist.csv', header=None)
df2 = pd.read_csv('out-degree_dist.csv', header=None)

#In-Degree Distribution
df1.plot(x=0, y=1, label="in-degree")
plt.grid(True)
plt.yscale('log')

axes = plt.gca()
axes.set_xlim([0,500])

plt.show()

#Out-Degree Distribution
df2.plot(x=0, y=1, label="out-degree")
plt.grid(True)
plt.yscale('log')

axes = plt.gca()
axes.set_xlim([0,420])

plt.show()
```
